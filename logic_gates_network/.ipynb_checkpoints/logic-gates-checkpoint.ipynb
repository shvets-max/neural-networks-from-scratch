```python
import numpy as np
import matplotlib.pyplot as plt
from typing import Tuple
from itertools import product
```

### Neuron class

```python
class Neuron:
    """Neuron class with simple activation"""
    def __init__(self, x: np.ndarray, weights: np.ndarray, thresh: float, outputs: Tuple[int|float, int|float] = (0, 1)):
        self.x = x
        self.weights = weights
        self.thresh = thresh
        self.outputs = outputs

    def calculate(self):
        net = np.dot(self.x, self.weights)
        return self.outputs[0] if net < self.thresh else self.outputs[1]
```

```python
def plot_results(X, title='title'):
    xpoints_0 = np.array([x[0] for x in X if x[2] == 0])
    ypoints_0 = np.array([x[1] for x in X if x[2] == 0])

    xpoints_1 = np.array([x[0] for x in X if x[2] == 1])
    ypoints_1 = np.array([x[1] for x in X if x[2] == 1])

    plt.plot(xpoints_0, ypoints_0, 'o', color='blue', markersize=12)
    plt.plot(xpoints_1, ypoints_1, 'o', color='red', markersize=12)
    plt.xlabel("x1", color='black', size=15)
    plt.ylabel("x2", color='black', size=15)
    plt.title(title, color='black', size=15)
    plt.show()
```

```python
prods = list(product([0, 1], [0, 1]))
prods
```

### OR architecture

```python
def _or(x1, x2):
    neuron_1 = Neuron(x=[x1, x2], weights=[1, 1], thresh=1.0)
    return neuron_1.calculate()
```

```python
[_or(p[0], p[1]) for p in prods]
```

```python
plot_data = [(p[0], p[1], _or(p[0], p[1])) for p in prods]
plot_results(plot_data, "OR")
```

### AND architecture

```python
np.dot([1,1], [1,1])
```

```python
def _and(x1, x2):
    neuron_1 = Neuron(x=[x1, x2], weights=[1, 1], thresh=2.0)
    return neuron_1.calculate()
```

```python
[_and(p[0], p[1]) for p in prods]
```

```python
plot_data = [(p[0], p[1], _and(p[0], p[1])) for p in prods]
plot_results(plot_data, "AND")
```

### NOR architecture

```python
def nor(x1, x2):
    neuron_1 = Neuron(x=[x1, x2], weights=[1, 1], thresh=1.0)
    y_1 = neuron_1.calculate()
    
    neuron_2 = Neuron(x=[y_1], weights=[-1], thresh=0.0)
    y_2 = neuron_2.calculate()
    return y_2
```

```python
[nor(p[0], p[1]) for p in prods]
```

```python
plot_data = [(p[0], p[1], nor(p[0], p[1])) for p in prods]
plot_results(plot_data, "NOR")
```

### NAND architecture

```python
def nand(x1, x2):
    neuron_01 = Neuron(x=[x1], weights=[-1], thresh=0.0)
    y_01 = neuron_01.calculate()
    neuron_02 = Neuron(x=[x2], weights=[-1], thresh=0.0)
    y_02 = neuron_02.calculate()
    
    neuron_11 = Neuron(x=[y_01, y_02], weights=[1, 1], thresh=1.0)
    y_11 = neuron_11.calculate()
    return y_11
```

```python
[nand(p[0], p[1]) for p in prods]
```

```python
plot_data = [(p[0], p[1], nand(p[0], p[1])) for p in prods]
plot_results(plot_data, "NAND")
```

### Complex gates

```python
prods3 = list(product(*3*[[0, 1]]))
prods3
```

```python
def or_and(x1, x2, x3):
    res1 = _or(x1, x2)
    res2 = _and(res1, x3)
    return res2
```

```python
[or_and(*p) for p in prods3]
```

```python
prods4 = list(product(*4*[[0, 1]]))
prods4
```

```python
def nand_nor_and_4(x1, x2, x3, x4):
    nand_12 = nand(x1, x2)
    nor_34 = nor(x3, x4)
    res = _and(nand_12, nor_34)
    return res
```

```python
[nor_nand_4(*p) for p in prods4]
```
